##### Changing Credentials and Using Strong Passwords
***

After completing this topic, you should be able to: ``choose strong passwords``.

Virtually every user of technology these days would be familiar with the implementation of a password but despite their prevalence, they have been proven to be one of the main security vulnerabilities and this is often due to the fact that people will use simple passwords - might use children's names, birthdates, pet's names, or even their own name or part of it as the password or maybe part of the password but in many cases, if the attacker knows the person fairly well, they might have a fairly good chance of guessing the password. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Second Factor or Multi-factor Authentication
***

Can improve upon password security by implementing what's known as second factor, or multi-factor authentication, which just adds in more components over and above the password itself, so it begins with:

> something you know 

- that would include a password but it might also be a PIN or a pattern, or maybe even a gesture that you have to perform 

Then you can combine that maybe with:

> something you have 

- a physical component such as a smart card or a fob 

If you want to keep going, you can add in:

> something you are 

- a biometric feature such as a fingerprint or maybe a retina scan

So again it's the combination of them that make everything more secure - I might be able to guess your password, but if I also need to swipe your fingerprint, well, that certainly makes it more difficult. 

This is nothing really new, particularly if you think about bank cards or credit cards - you have to have the card itself and you must also have the PIN, so you need both, the something you have and the something you know - the PIN by itself is useless without the card - the card by itself is useless without the PIN, so it implements stronger authentication. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Strong Password Creation
***

With respect to just the password itself, you can still increase the security by implementing strong passwords and this involves avoiding personal information, for example, nothing that relates to you or your family and you should also avoid easily recognizable numbers, such as your address or phone number - particularly relates to a PIN. 

A lot of PINs on bank cards are only four digits, so people will often use the same four digits at the end of their phone number for the PIN, so again, if you know the person's phone number, that would likely be the first thing you might try. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Avoid Read Words
***

Can also avoid using real words - this makes it challenging for password cracking tools to guess the password. 

This is the dictionary attack - these types of programs simply run through every word in the dictionary. 

Certain programs can also process combinations of words or letters or numbers - that will take them more time, but they might still be able to gain access eventually, so use a combination that includes upper and lowercase letters with the addition of numbers and special characters and that will make it significantly more difficult for a dictionary attack to succeed. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Longer Passwords
***

Can also use longer passwords. 

A password that is at least eight characters in length is considered to be strong, particularly if you compare that to a password that is only three or four characters in length and they are quite simply harder to crack because of the higher number of characters, there are more possibilities. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Implement Passphrases
***

Might also be able to implement passphrases. 

A lot of applications and or operating system environments will accept up to 128 or maybe even 256 characters in the password, so you can basically just use these characters from a string of text, maybe a line from a song or a quote, or anything along those lines but again, you want to be a little bit careful with that, because if people know you very well, they might hear you use that phrase a lot in day to day conversation, so again, it might be easy for someone to guess what it is but it still is harder to guess, at least to something like a dictionary attack. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Password Storage
***

With respect to password storage, ideally you want users to remember the password or perhaps more specifically stated, you don't want them writing it down and then just hiding it under the keyboard or quite literally posting it on their desk or their monitor but there is a little bit of a balance here that you might want to try to find, because if you do require your users to create very complex passwords, then they will almost never remember it and conversely, they will almost always write it down.

If they can secure the written down copy in a locked drawer, for example, then that is more secure because that exceptionally difficult password is almost impossible to guess by someone else, and very secure against something like a dictionary attack, so again, you do get enhanced security by having that impossible to guess password, but you need the written down copy to be secured as well, so if it was an attacker trying to use a dictionary attack, well, you're very safe against those but you're not all that safe against someone on the inside who might have access to the drawer to get the written down copy - again, there's a little bit of a balance that you might have to strike there. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Password Management Tools
***

But there are password management tools, software and applications that store the username and password and these usually do so in encrypted format, so that users don't have to remember it - the application itself will remember it and a lot of people might be familiar with the fact that browsers actually remember usernames and passwords. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Password Changing
***

Another aspect to consider is changing the password, so you should instruct users to update passwords on a regular basis, or perhaps more commonly, implement policies that require them to do so. 

The idea here being that if they have the same password forever, sooner or later somebody will figure out what it is, so by changing it, you force them to start all over again. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Passwords for Different Accounts
***

You should also use different passwords for different accounts, in other words, if you use the same password for everything, then you are certainly a lot more exposed to losses if somebody does get your password because they'd be able to access everything that you are securing with that password. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Not Save Password in the Browser
***

Generally, do not save password information in the browser - that comes back to what I just mentioned about the password management applications - there are some known as vaults that are dedicated to this process, but browsers do it as well and that might be okay as long as you are in a private environment, such as your home or your office at work, where you are the only one who uses that computer but certainly, if it's a public place, such as an Internet cafe, a library, or a hotel lobby, you certainly should not save password information in the browser because somebody who might just be doing a little bit of shoulder surfing could come right in behind you and access that same site that you just accessed with your account information.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Password Implementation that meet your needs
***

It might take a little bit of thought, a little bit of balance but ultimately, it's up to you to try to find a password implementation that is secure enough to meet your needs while not being so difficult to remember that you write it down and stick it on your monitor and hence defeat the purpose of password security.

###### Hardware Trust (Firmware Trust) and Updating Firmware
***

After completing this topic, you should be able to: ``update firmware``.

In this presentation, we'll cover hardware trust, or perhaps more specifically stated, firmware trust and say that because, these days, a lot of hardware will come with an embedded chip right in the device and that chip contains a certain amount of software, dedicated instructions for how that particular component performs its job and that software is known as firmware. 

Given that it is dedicated software for a specific device that performs a specific task, they are usually very reliable - don't have to do all the things that an operating system has to do, for example but at the end of the day, it is still software that has been programmed for this hardware, so the software itself is still vulnerable to a degree. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Trusted Platform Module (TPM)
***

The ``Trusted Platform Module``, or the ``TPM`` is an embedded chip that is usually found these days in laptops, desktops, and servers, even other devices such as set-top boxes, like a PVR, or your TV tuner, smartphones, or even tablets and it's a device that was defined in the ``National Institute of Standards and Technology`` - ``NIST`` - trusted computing requirements and the SP stands for special publication, and then those numbers just indicate which document:


- SP 800-147


- SP 800-155


- SP 800-164 


The idea behind a TPM is that it stores cryptographic keys - these, of course, can be used for encryption purposes to protect information. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Trusted Platform Module (TPM) - Microsoft BitLocker
***

A specific application of this is Microsoft BitLocker key storage - designed to encrypt all of the data on a volume so that everything is protected, not just individual files. 

If you don't have a TPM, some older systems might not, you can still implement BitLocker by using a USB flash drive - it can store those keys on the flash drive, but that's not as secure because the keys are readily available, if you will, on the flash drive whereas with a TPM, they are embedded right into the system. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Trusted Platform Module (TPM) - Initializes creates a Storage Root Key (SRK)
***

The TPM itself, when it initializes, it creates what's known as a ``Storage Root Key`` (``SRK``) and then any given application such as BitLocker can use this root key to protect its own application-based keys.

This is a fairly common approach when it comes to applications that use encryption keys. 

A lot of database service, for example will create a key that resides at the server level or the service level because any one database server can have any number of databases on that server, so there's this master server level key, if you will - then for each individual database that I want to protect, I can create different database level keys, so the same thing goes with this TPM root key. 

That is, in essence, the master key and then any other application that needs to protect information can use that root key to protect its own application-based key, so BitLocker is an example of an application-based key. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Trusted Platform Module (TPM) - Detect Boot Changes
***

The TPM can also detect boot changes, and as such it can mitigate boot-kit threats - master boot record rootkits and the [2011 Mebromi](https://www.webroot.com/blog/2011/09/13/mebromi-the-first-bios-rootkit-in-the-wild/) attack are examples of these boot-kit threats.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Rootkit
***

Boot-kit and rootkit, are fairly similar but for the most part, a rootkit is a virus that gets into the system that tries to hide in plain sight, if you will, by disguising itself as legitimate files. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Bootkits
***

Boot-kits usually extend the capability of rootkits so that they can survive reboots, so if you just try to get rid of something because it's active in memory by rebooting - a boot-kit basically gets itself into the boot record and survives these reboots.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### TPM - Change to the Boot Environment is Detected
***

With the TPM, any kind of change to the boot environment is detected as suspicious activity and the system will then reboot and it will require a recovery key password, which, in essence, resets the boot environment back to the way it was and the protection that was implemented is able to say, well, I know that this is what it did look like before the suspicious activity, so I can just discard whatever change occurred and set everything back to the way it was, so, again, that's the idea behind the TPM, it can prevent that tampering of the boot environment again you just configure something like a password to ensure that you are an authorized user and then, in essence, your boot environment is back to the way it was. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### BitLocker
***

With respect to BitLocker, this particular application was designed to provide entire disk volume encryption.

- ``EFS``, the ``Encrypting File System`` is a feature of Windows and the ``NTFS`` file system but that is used for individual files and folders

This is used for the entire volume, so fixed disks on your system can be protected at the disc level and basically, this is designed to protect data at rest, so that if someone steals your system, even if they attempt to gain access to the data on the drive by removing the drive and attaching it to their own system where they are a local administrator, they cannot remove the encryption.

They would be able to remove the NTFS permissions in that manner but the encryption is still there and requires the storage root key of the TPM to be able to access the data, so by just removing the hard drive, they don't get anywhere and, of course, even if they leave it in the system, because they stole your whole laptop, they would still need the password or the PIN to boot the system because you can configure the TPM to require a PIN, so as long as they do not have that PIN, they cannot gain access to the data.

It can also be implemented for removable media, so if you have an external USB hard drive or a flash drive, you can implement ``BitLocker To Go`` to protect the data on those drives as well and that's fairly common, of course, because they're very easily lost or stolen, so, again, the idea is to protect the entire contents of that disc or volume.

###### Firmware Updates
***

Firmware updates, of course, are very common - computers contain a lot of firmware - USB peripherals, graphic cards, web cams - malware can be injected into the firmware of computers, so you want to regularly check with the manufacturer for firmware updates, and some devices will often inform you - just pop up and tell you there is an update, but not always, so it's a good idea to keep on top of it and to make sure that everything is current and usually the updates are security related - they'll say, we found a security vulnerability, here's the fix for it, so please apply this.

###### Conclusion - TPMs help to keep System more Secure

But ultimately, it's like anything, you just want to keep on top of it and ensure the standard activity that you do on a day-to-day basis does not invite these types of malware applications but, again, some of them are fairly tricky to detect and even to remove, so TPMs can certainly help you in keeping your system more secure.

Then, for lack of a better description, the newer version might be just a complete roll up of all of the earlier patches plus perhaps some new features, of course but the idea in most cases is that vulnerabilities are discovered over time and they can be related to security or stability, or really anything, compatibility perhaps, so patching fixes those problems in the short term, if you will, until the next version.

###### File Hashing
***

After completing this topic, you should be able to: ``implement file hashing``.

The concept of hashing, at a very simple level, it is taking a value, running it through some kind of mathematical algorithm or function to produce another value.

It's similar to encryption, but it doesn't really have the same approach - if, for example, I have a document that I want to ensure that no one can read, then I might encrypt it, that renders the data unreadable to anyone who does not have the decryption key. 

Hashing is not so much concerned with rendering it unreadable but rather ensuring that the information does not change, so for a very simple example, imagine you have a document and you open it up and you type in the number two - then you say, I want to apply a hash algorithm to that file, and the function or the algorithm is to simply say, take whatever the value of the document was and add two to the value, so it should then read the value of two, add two to the value and come up with an answer of four - four represents the hash value. 

If then somebody opens the document without me knowing it and alters the data and changes it to three, then they close down the file and say, all right, I've compromised this file now, the hash will change because the hash said to simply take the existing value and add two to it, so now I get a hash value of five, and I know that this document has changed - again, that's a very basic example, but that is the essence of hashing - take some kind of information, run it through a function, produce a hash value, if that value changes, then the data has to have changed. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Algorithms - ``MD4``
***

MD4 was invented by Ron Rivest - that is the R of RSA encryption used in PKI, Public Key Infrastructure. 


- ``MD`` - stands for Message Digest


- ``4`` - the version


This algorithm produces a 128-bit hash value, which is 16 bytes.

The security of MD4 itself has been compromised, in other words, an attacker can modify the contents of the original file and then compromise the hash, so that the hash still comes out to be the same even though the original data did change, so it's not recommended to be used any longer. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Algorithms - ``MD5``
***

MD5 was a newer version, still a Cryptographic hash function, and it still produced a 128-bit hash value, which is 16 bytes, so the hash value, the answer is represented by a 32 digit hexadecimal number and again, the common uses were for applications and maintaining the data integrity. 

Back to my example, obviously entering in the number two would be very easily detected if someone changed it to three but for very large documents, in particular, databases, a small change can be very difficult to detect but hashing can always detect it so we can maintain the data integrity by using hashing. 

MD5, of course, replaced the earlier MD4 function, but MD5 has since also been compromised, what's known as a ``Collision attack`` and or a ``Rainbow table`` can be used for ``Password cracking`` so that when you hash passwords - a Rainbow table, in essence, refers to all of the colors of the rainbow, it tries every possibility, so MD5 also no longer recommended for production use. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Algorithms - ``SHA``
***

SHA is the Secure Hash Algorithm and this is a family of cryptography hash functions and it was designed by the National Institute of Standards and Technology, as well as the National Security Agency, to effectively be the successor. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Algorithms - ``SHA-0``
***

SHA-0 of course was the first release and this was quite some time ago, so significant flaws were identified - produced 160-bit hash function, but it had a maximum message size of 2 to the power 64 minus 1 bits, and it was of course replaced by SHA-1. 

< [Table of Contents](#top) | [References](#references) >

***
###### Algorithms - ``SHA-1``
***

SHA-1 came along, and it still used 160-bit hash function - was designed by the NSA, the National Security Agency, and this is a ``Digital Signature Algorithm`` with a maximum message size, again of 2 to the power of 64 minus 1 but it was also identified as susceptible so it was no longer approved after 2010 for most cryptographic uses.  

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Algorithms - ``SHA-2``
***

SHA-2 consists of both 256 and 512-bit algorithms, which produce 32 or 64-bit words - a word is just a unit of measure for how many characters are produced by the hashing algorithm.

There are some truncated versions as well, SHA-224, 384, 512 with 224 and 256, so these are just variations on how many bits are sent in to the algorithm and how long the resulting hash value is. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Algorithms - Keccak - ``SHA-3``
***

The newest version is SHA-3, formally known as ``Keccak``, and this was released by NIST on August 5th, 2015, so this is approved, of course, for current use. 

It uses what's known as a ``Sponge construction``, meaning that it effectively soaks up the data, if you will - put a sponge in water, it soaks up the water and when you squeeze it you ring out the water, but it's kind of in a different form if you will, so that's just a loose analogy of how the data is absorbed, and then how it comes out of the algorithm. 

With SHA-3, there is no limit on the size of the message, so it can basically be used against anything regardless of how large it is - now, the earlier ones were still quite large, you know, 2 to the power of 64 is enormous, but is still a limit, so with SHA-3, you can run anything through in terms of the size, and you will still get a hashed version of that data coming out the other side.

###### Cryptography Terminology

When it comes to terminology, you might hear a fair number of these used somewhat interchangeably but for official definitions:

> Cryptanalysis is quite simply the study of Cryptosystems. 

> Cryptography itself is the use of tools and mechanisms designed to protect information, in essence, engineering solutions to come up with means to encrypt data. 

> Cryptology is how the mathematics apply to Cryptographic algorithms, because it is all based on math. 

> Cryptosystem is the collective overall implementation of all components.

> Algorithm, in and of itself, is just a mathematical function - can send a number or a value into the algorithm and it will produce something coming out on the other side. 

> Cipher is quite simply an example of an algorithm, but one that is specifically designed for encryption purposes. 

> Plaintext or cleartext, is essentially the data before it is protected, so the plain message that you are wanting to send, for example. 

> Ciphertext or cryptogram refers to the result of encrypting the cleartext, so it's what you see on the other side of the algorithm - the function produces the Ciphertext. 

> Encipher or Encrypt, or Encode means to apply the encryption method and this is often achieved with whats known as a public key, and that refers back to the PKI - that Public Key Infrastructure.

> Key is the value or the parameter applied to the seed material, in essence, the algorithm.

> Key space is the total number of available keys within a Cryptosystem, and the more keys you have, the stronger the system.

> Initialization Vector ``IV`` is a random value that's used when encrypting data, so that if you do happen to encrypt the exact same plaintext, the results will vary - in other words, this is kind of the opposite of key clustering - you don't want the same value to produce the same encrypted data - want it to produce different data.

A lot of passwords will support up to 128 characters, so in some cases you might even implement a pass phrase of some kind, but that is certainly your call but the general rule of thumb is, of course, that shorter passwords that don't mix the case and don't mix the letters are certainly going to be easier to guess, or for some sort of password-cracking program to be able to ascertain. 

ii. Then we asked to describe file hashing. 

This is the process of applying a mathematical algorithm to a file to produce a value that can be used to determine if the file contents have changed, so again, you start with the original value, which is the contents of the file itself and then you use that information to run it through some kind of a function that produces an answer, for lack of a better word, so then if the original data changes, and you run it through the same function, you will get a different hash value - a different answer will come out if you start with a different file, therefore, this indicates that the file has been changed, so if you know it should not have been changed, then in essence, somebody has tampered with that file. 

iii. Then we asked to describe service hardening, and this is simply the process of disabling services on a server that are not in use, so if you go to just about any server and examine the services that are running, in many cases, there are more running than what is necessary. 

If that server is being dedicated to a specific task, for example, then the tasks that it does not perform should simply be disabled because each running service has a port associated with it and if that service is running, then the port is open, and it presents an avenue of attack, so quite simply, shutting down that service disables the port. 

iv. Then describe device hardening and this is the same approach, but more so referring to disabling physical ports that are not in use, such as on routers or switches or blocking ports on a firewall. 

This protects the entire network, in many cases, by simply not allowing any kind of communication over that specific port. 

In both cases, hardening just refers to shutting down that which is not in use or undesirable and the more you reduce those open ports, the more you reduce the attack surface of your overall environment.
